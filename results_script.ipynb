```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import os
os.getcwd()
import math
import numpy as np
np.sqrt([2,2])
```

```python
data = pd.read_csv('simulations_results.csv')  #get direction of simulation files!
data.head(20000/50+2)
results = list(data.columns)[2:]
data_env=pd.read_csv('simulations_env_shift_results.csv')
results
```

```python
colors=['b','r','g','y','k','darkgreen', 'darkblue','orange', 'pink', 'brown', 'magenta', 'darkviolet']
```

```python
plt.figure(1, figsize=(15,10))
#plt.ylim(0, 1)
plt.plot(color="blue", linewidth=2.5, linestyle="-", label="cosine")


plt.legend(loc='upper left', frameon=False)
for i in xrange(12):
    plt.plot(np.sqrt(data[results[4]][i*400+i:(i+1)*400+i]), colors[i], label=data['Unnamed: 0'][i*400+i])
    plt.plot(np.sqrt(data[results[2]][i*400+i:(i+1)*400+i-200]), colors[i])
    plt.legend(loc='upper left', frameon=False)
   
```
