```python
data = pd.read_csv('Downloads/simulations_results.csv')  #get direction of simulation files!
data.head(20000/50+2)
results = list(data.columns)[2:]
data_env=pd.read_csv('Downloads/simulations_env_shift_results.csv')
results
```
